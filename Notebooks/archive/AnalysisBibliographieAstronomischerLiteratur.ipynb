```python
import json
import pandas as pd
```

```python
with open('BiblioAstroLib.json') as jsonData:
    BibAstroLit = json.load(jsonData)
```

```python
BibAstroLit.keys()
```

```python
df = pd.DataFrame.from_dict([BibAstroLit])
```

```python
user_ids = []

frames = []

for user_id, d in BibAstroLit.items():
    user_ids.append(user_id)
    if d:
        frames.append(pd.DataFrame.from_dict(d, orient='index'))

df = pd.concat(frames, keys=user_ids)
df.columns = ['fulltext']
df.index.rename(['Letter','Name'], inplace=True)
df.sort_index()
```
